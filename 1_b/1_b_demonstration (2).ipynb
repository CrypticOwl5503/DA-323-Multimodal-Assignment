```python
import os
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
from collections import Counter
nltk.download('punkt')
nltk.download('stopwords')
```

```python
texts_folder = '/content/texts'


def process_text_file(file_path):
  with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read()
  return text
```

```python
def analyze_text(text, topic_name):
  # Tokenize the text
  tokens = word_tokenize(text)


  stop_words = set(stopwords.words('english'))
  filtered_tokens = [w.lower() for w in tokens if w.isalnum() and w.lower() not in stop_words]


  word_counts = Counter(filtered_tokens)
  most_common_words = word_counts.most_common(10)



  plt.figure(figsize=(10, 5))
  plt.bar(*zip(*most_common_words))
  plt.title(f'Top 10 Most Frequent Words in "{topic_name}"')
  plt.xlabel('Words')
  plt.ylabel('Frequency')
  plt.xticks(rotation=45, ha='right')
  plt.show()


  wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_counts)
  plt.figure(figsize=(10, 5))
  plt.imshow(wordcloud, interpolation='bilinear')
  plt.axis("off")
  plt.title(f'Word Cloud for "{topic_name}"')
  plt.show()
```

## Sentiment analysis and word length count

```python
import os
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
from textblob import TextBlob

!unzip /content/texts.zip
nltk.download('punkt')
nltk.download('stopwords')

texts_folder = '/content/texts'

def process_text_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

def analyze_text(text, topic_name):
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [w.lower() for w in tokens if w.isalnum() and w.lower() not in stop_words]
    word_counts = Counter(filtered_tokens)
    blob = TextBlob(text)
    sentiment_polarity = blob.sentiment.polarity
    sentiment_subjectivity = blob.sentiment.subjectivity
    text_length = len(text)
    print(f"Topic: {topic_name}")
    print(f"Sentiment Polarity: {sentiment_polarity}")
    print(f"Sentiment Subjectivity: {sentiment_subjectivity}")
    print(f"Text Length: {text_length}")
    plt.figure(figsize=(8, 4))
    plt.bar(['Polarity', 'Subjectivity'], [sentiment_polarity, sentiment_subjectivity], color=['blue', 'orange'])
    plt.title(f'Sentiment Analysis - {topic_name}')
    plt.show()
    plt.figure(figsize=(8, 4))
    plt.hist([len(sentence) for sentence in text.split('.')], bins=30, color='purple', edgecolor='black')
    plt.title(f'Text Length Distribution - {topic_name}')
    plt.xlabel('Sentence Length')
    plt.ylabel('Frequency')
    plt.show()

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

for filename in os.listdir(texts_folder):
    if filename.endswith('.txt'):
        topic_name = filename[:-4]
        file_path = os.path.join(texts_folder, filename)
        text = process_text_file(file_path)
        analyze_text(text, topic_name)
```
